```python
import pandas as pd

# Google. (n.d.). Gemini. Google. https://gemini.google.com/ I used Artificial Intelligience for assistance in displaying the data.

file_path = '/content/extra_credit_data.xlsx'

try:
    df = pd.read_excel(file_path)
    print(f"Successfully loaded data from {file_path}.")
    display(df.head())
except FileNotFoundError:
    print(f"Error: The file at '{file_path}' was not found. Please check the path and try again.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
print("First 5 rows of the DataFrame:")
display(df.head())

print("\nDataFrame Info:")
df.info()

print("\nDescriptive Statistics:")
display(df.describe(include='all'))
```

```python
import numpy as np

# Google. (n.d.). Gemini. Google. https://gemini.google.com/ I used Artificial Intelligience for assistance in cleaning the data.


# Define special missing values
special_missing_values = ['.i: Inapplicable', '-70']

# Columns to apply the replacement
columns_to_clean = ['agewed', 'divorce', 'spwrkslf', 'hapmar', 'ballot', 'childs', 'sppres80']

# Replace special missing values with np.nan
df[columns_to_clean] = df[columns_to_clean].replace(special_missing_values, np.nan)

print("Missing values after converting special values to np.nan:")
print(df.isnull().sum())
```

```python
import numpy as np

# Google. (n.d.). Gemini. Google. https://gemini.google.com/ I used Artificial Intelligience for assistance in determining categorical columns.

# Define categorical columns
categorical_cols = ['marital', 'divorce', 'spwrkslf', 'hapmar', 'ballot']

print("Converting categorical columns and verifying:")
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')
        print(f"\nColumn '{col}':")
        print(f"  Data Type: {df[col].dtype}")
        print(f"  Unique Values (excluding NaN): {df[col].dropna().unique().tolist()}")
    else:
        print(f"Column '{col}' not found in DataFrame.")
```

```python
print("Descriptive statistics for 'agewed', 'sppres80', and 'childs' columns:")
display(df[['agewed', 'sppres80', 'childs']].describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Google. (n.d.). Gemini. Google. https://gemini.google.com/ I used Artificial Intelligience for assistance in plotting the data.

# Numeric columns to visualize
numeric_cols = ['agewed', 'sppres80', 'childs']

# Create a figure with subplots
plt.figure(figsize=(18, 5))

# Plotting agewed
plt.subplot(1, 3, 1)
sns.histplot(df['agewed'].dropna(), kde=True)
plt.title('Distribution of Age When First Married')
plt.xlabel('Age When First Married')
plt.ylabel('Frequency')

# Plotting sppres80
plt.subplot(1, 3, 2)
sns.histplot(df['sppres80'].dropna(), kde=True)
plt.title('Distribution of Spouse Occupational Prestige Score (1980)')
plt.xlabel('Spouse Occupational Prestige Score')
plt.ylabel('Frequency')

# Plotting childs
plt.subplot(1, 3, 3)
sns.histplot(df['childs'].dropna(), kde=True)
plt.title('Distribution of Number of Children')
plt.xlabel('Number of Children')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Google. (n.d.). Gemini. Google. https://gemini.google.com/ I used Artificial Intelligience for assistance in creating the confusion matrix.


# 1. Select the numeric columns
numeric_cols_for_corr = ['agewed', 'sppres80', 'childs', 'year']

# 2. Calculate the correlation matrix
correlation_matrix = df[numeric_cols_for_corr].corr()

# 4. Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

# 5. Add a descriptive title to the heatmap
plt.title('Correlation Matrix of Numeric Variables')

# 6. Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Google. (n.d.). Gemini. Google. https://gemini.google.com/ I used Artificial Intelligience for assistance in creating the boxplots.


plt.figure(figsize=(12, 7))
sns.boxplot(data=df, x='marital', y='agewed')
plt.title('Age When First Married by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Age When First Married')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Google. (n.d.). Gemini. Google. https://gemini.google.com/ I used Artificial Intelligience for assistance in crosstabulating the variables.

# 1. Create a crosstabulation of 'marital' and 'divorce' columns
crosstab_marital_divorce = pd.crosstab(df['marital'], df['divorce'])

# 2. Create a heatmap of the resulting crosstabulation table
plt.figure(figsize=(12, 8))
sns.heatmap(crosstab_marital_divorce, annot=True, fmt="d", cmap='viridis', linewidths=.5)

# 3. Add a descriptive title to the heatmap
plt.title('Relationship between Marital Status and Divorce Status')

# 4. Label the x-axis as 'Divorce Status' and the y-axis as 'Marital Status'
plt.xlabel('Divorce Status')
plt.ylabel('Marital Status')

# 5. Display the plot
plt.show()
```
